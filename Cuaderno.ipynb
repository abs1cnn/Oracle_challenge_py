### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unimos todas las tiendas en un solo DataFrame
datos = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

#1. Análisis de facturación



```python
datos['Total Venta'] = datos['Precio'] + datos['Costo de envío']
facturacion = datos.groupby('Tienda')['Total Venta'].sum().sort_values()

import matplotlib.pyplot as plt

facturacion.plot(kind='barh', color='mediumseagreen')
plt.title('🧾 Facturación Total por Tienda')
plt.xlabel('Total Vendido (COP)')
plt.ylabel('Tienda')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
ventas_categoria = datos['Categoría del Producto'].value_counts()

ventas_categoria.plot(kind='bar', color='cornflowerblue', figsize=(10, 5))
plt.title('📦 Cantidad de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificaciones = datos.groupby('Tienda')['Calificación'].mean().sort_values()

calificaciones.plot(kind='bar', color='orange')
plt.title('⭐ Calificación Promedio por Tienda')
plt.ylabel('Calificación')
plt.ylim(0, 5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos = datos['Producto'].value_counts().head(5)
productos_menos_vendidos = datos['Producto'].value_counts().tail(5)

print("✅ Productos más vendidos:")
print(productos_mas_vendidos)

print("\n⚠️ Productos menos vendidos:")
print(productos_menos_vendidos)
```

# 5. Envío promedio por tienda

```python
envio_promedio = datos.groupby('Tienda')['Costo de envío'].mean().sort_values()

envio_promedio.plot(kind='bar', color='salmon')
plt.title('📦 Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de envío')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
